```python
import os, sys
import requests
from bs4 import BeautifulSoup
import pandas as pd
import datetime
import re
import difflib
import random
import time
```

```python
headers = {'user-agent': 'Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.88 Mobile Safari/537.36'}
today = datetime.date.today()

regex = '\d{4}'
patterm = re.compile(regex)

StockNo = '1101'
StockData = []

columns = ['日期', 
           '外陸資買進股數(不含外資自營商)', '外陸資賣出股數(不含外資自營商)', '外陸資買賣超股數(不含外資自營商)', 
          '外資自營商買進股數', '外資自營商賣出股數', '外資自營商買賣超股數', 
          '投信買進股數', '投信賣出股數', '投信買賣超股數',
          '自營商買賣超股數', '自營商買進股數(自行買賣)', '自營商賣出股數(自行買賣)', '自營商買賣超股數(自行買賣)', 
          '自營商買進股數(避險)', '自營商賣出股數(避險)', '自營商買賣超股數(避險)',
          '三大法人買賣超股數']

CountDay = 0
CountDayTrade = 0
TradingDayStr = []
url = []
DateLengh = 60

while CountDayTrade < DateLengh:    
    TradingDay = today - datetime.timedelta(days=CountDay)    
    urlTemp = "https://www.twse.com.tw/fund/T86?response=html&date="+TradingDay.strftime('%Y%m%d')+"&selectType=01"

    r = requests.get(urlTemp, headers=headers)
    soup = BeautifulSoup(r.text, 'html5lib')
    
    soup.find(name='div')
    s1 = soup.get_text()
    s2 = "很抱歉，沒有符合條件的資料!"    
    matchratio = difflib.SequenceMatcher(None, s1, s2).ratio()
    if matchratio < 0.03:
        TradingDayStr.append(TradingDay.strftime('%Y/%m/%d'))        
        url.append(urlTemp)        
        
        AllData =[]
        DataTemp = [] 
        
        for i in soup.find(name='tbody').find_all(name='td'): 
            AllData.append(i.get_text())
            
        for i in range(len(AllData)):
            if AllData[i] == StockNo:
                for j in range(2,19):
                    if j == 2:    
                        DataTemp.append(TradingDayStr[-1]) 
                        DataTemp.append(AllData[i+j]) 
                    else:
                        DataTemp.append(AllData[i+j]) 
                StockData.extend([DataTemp])        
        
        CountDay = CountDay + 1
        CountDayTrade = CountDayTrade + 1
        
    else:
        CountDay = CountDay + 1
    time.sleep(3)

print("股票代號", StockNo)
df = pd.DataFrame(StockData, columns=columns)
df   
```
