## requests + BeautifulSoup


```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.dcard.tw/f'
r = requests.get(url)
r.encoding = 'utf-8'
print(type(r.text), '\n', r.text)
```

```python
soup = BeautifulSoup(r.text, 'html5lib')
print(type(soup), '\n', soup)
```

## Grab + PyQuery

```python
from grab import Grab

url = 'https://www.dcard.tw/f'

g = Grab()
resp = g.go(url)
print(type(resp.body), '\n', resp.body)
```

```python
from pyquery import PyQuery as pq

doc = pq(resp.body)
print(type(doc), '\n', doc)
```
